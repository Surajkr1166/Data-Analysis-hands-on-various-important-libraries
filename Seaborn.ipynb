##1.1 displot()/histplot()

- it will work for only one col
- it shows how many times range appears in your data.
- It splits the data into small groups (called bins) and shows how many items are in each group.
- only one colname we need to pass if u pass more colnames it will throw the error

```
syntax:-
    sns.displot(x='colName' / y='colName' , data=df_var)
    sns.histplot(x='colName' / y='colName' , data=df_var)
```

###Attributes:-

####1. kde(kernal density estimation)
- A way to show how data points are spread out, by creating a smooth curve that shows the most common values.

####2. bins
- A range of values divided into equal parts , used to group and organize data for analysis in easy way.
- ```Syntax--> no.bins <= no.unique value```

####3. hue
- hue it is used to differentiate every unique value with different colour
- ```Syntax --> hue='cat_col'```

####4. palettete
- palettete is used to change the colors of hue attribute
- ```Syntax  --> palette = 'VAL'```
- VAL = 'Set1/Set2/Set3/Coolworm/rainbow/twilight/magma/Reds/Greens/Blues' and 'dark:colName','light:colName'

####5. multiple='dodge'
- it is used to place the bars side by side

####6. shrink=0.0 - 0.9
- it is used to provide the space b/w the bar

##2.4 Box plot()
- it is used to identify the outliers in a particular col

###Outlier
- it is a data point ,that is very far from the other data points or irrelated from the other data points.
- four quartiles-->
- - 0% to 25%-->1st quartile
- - 25% to 50%-->2nd quartile
- - 50% to 75%-->3rd quartile
- - 75% to 100%-->4th quartile
- **NOTE**--> if any value is far from the mean val then it is a outlier
```
syntax:-sns.boxplot(x='catcol',y='num_col',data=df_var,hue='cat_col' )
```


### Before Handling OutLiers

`````python
sns.boxplot(x='pH' , data=dfw )
`````

### Handling Outliers Example

`````python
Q1 = dfw['pH'].quantile(0.25)  #quantile is used to detect outliers
Q3 = dfw['pH'].quantile(0.75)
print(Q1,Q3)
`````

`````python
IQR = Q3-Q1        #internal quartile range (iqr)

lb = Q1-1.5*IQR    #lower boundary  (here 1 is a standard constant value to get outliers / 1.5 is balanced)
ub = Q3+1.5*IQR    #upper boundary

print(lb , ub)
`````

`````python
Outlaier  =  dfw['pH'][(dfw['pH']<lb) | (dfw['pH']>ub)]    #Fetching Outliers
print(Outlaier)
`````

`````python
Otlfree = dfw[(dfw['pH']>lb) & (dfw['pH']<ub)]  #Fetching InnerLiers
med = Otlfree['pH'].median()
print(med)
`````

`````python
dfw.loc[Outlaier.index ,'pH'] =  med     #Filling Outliers with Median
dfw.loc[Outlaier.index ,'pH']
`````

### After Handling Outliers

`````python
sns.boxplot(x='pH' , data=dfw )
`````

## Differences between box plot and violin plot

#### boxplot

- 1)it displays mean and quartiles

- 2)it will shows outliers

- 3)it does not show density

- 4)it is simple and easy to read

####violinplot

- 1)it displays density

- 2)does not show outliers explicitly. Individual points it wont display but its involved in the shape

- 3)More informative than box plots

- 4)it can be more difficult to read. If the people does not have knowledge on density plot
