# Groupby Opration

To work with some particular value then we need to go for groupby operations.

It is a function which splits the data to do some operations.

syntax:

> ***1. gr_var = df_var.groupby('cat_col')   --------> single columns***

> ***2. gr_var.groups  ------>return dictionary(optional)***

> ***3. val_gr = gr_var.get_group('unique_val')***

> ***4. val_gr['unique_val'].agrregtion_function***

```python
import numpy as np
import pandas as pd
```

```python
dummy={'empName':['a','b','c','b','e','f'],'dept':['it','sales','devops','it','it','sales'],'sal':[20000,15000,35000,40000,50000,25000]}
emp=pd.DataFrame(dummy)
emp
```

```python
grb=emp.groupby('dept')
grb.groups
```

```python
vl = grb.get_group('it')
vl
```

```python
vl['sal'].sum()
```

### Multiple Columns

syntax:

> ***1. gr_var = df_var.groupby(['cat_col1','cat_col2',...,'cat_colN'])***

> ***2. gr_var.group()   ->return dictionary ---optional***

> ***3. value = gr_var.get_group(['cat_col1_unique_val','cat_col2_unique_val',...,'cat_colN_unique_val'])***

> ***4. value[['cat_col1_unique_val','cat_col2_unique_val',...,'cat_colN_unique_val']].agrregtion_function***



```python
dft =pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/titanic.csv ', encoding ='ISO-8859-1')
dft
```

```python
egr = dft.groupby(['sex','pclass'])
egr.groups.keys()
```

```python
fegr = egr.get_group(('female', 1))
fegr[['age','fare']].min()
```

### Optimation code (multiple columns)

syntax:

> ***output = df_var.groupby(['cat_col1','cat_col2',...,'cat_colN']).get_group(('cat_col1_unique_val','cat_col2_unique_val',...,'cat_colN_unique_val'))[['colName1','colName2',....,'colNameN']]***

```python
avg1 = dft.groupby(['sex','pclass']).get_group(('female', 1))[['age','fare']].mean()
avg1
```

# Concatination

syntax:-

> ***pd.concat([df_var1, df_var2,....., df_varN] , axis = 0)  ------>vertical concatination***

> ***pd.concat([df_var1, df_var2,....., df_varN] , axis = 1)   ------>horizontal concatination***

> ***pd.merge(df_var1, df_var2, on = comman_Col, how = 'type_of_the_join')***

> ***pd.merge(df_var1, df_var2, on = comman_Col, how = 'type_of_the_join')***

### 5. joining (more than 2 columns)

###syntax:-

> ***pd.merge(df_var1, df_var2, on = comman_Col, how = 'type_of_the_join')merge(df_var3, on = comman_Col, how = 'type_of_the_join')***

```python
dummy={'dept_id':[1,2,3,4],'BGroup':['A+','O+','O-','AB-']}
BGroup=pd.DataFrame(dummy)
BGroup
```

# Pivot_Table(Multiple aggregation function on multiple columns)

It is the power house of Pandas.

It is used to perform multiple agg. function on the multiple columns of the dataset.

It return as a DataFrame.

It only perfom on numerical columns.

****1. Single aggregation function on multiple columns****

syntax:

>***df_var[['colName1','colName2',...,'colNameN']].aggfunc()***


****2. Multiple aggregation function on multiple columns****

syntax:

> ***df_var.pivot_table(values=['colName1','colName2',...,'colNameN'],index=['cat_colName1','cat_colName2',...,'cat_colNameN'],aggfunc=['func1','func2',...,'funcN'])***

****3. Different aggregation function on different columns****

syntax:

>***df_var.agg({'colName1':['func1','func2',...,'funcN'],'colName2':['func1','func2',...,'funcN'],...,'colNameN':['func1','func2',...,'funcN']})***

```python
dft =pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/titanic.csv ', encoding ='ISO-8859-1')
dft
```

```python
dft.pivot_table(values=['age','fare'],index='sex',aggfunc=['min','max','mean'])
```

```python
dft.pivot_table(values=['age','fare'],index=['sex','pclass'],aggfunc=['min','max','mean'])
```

```python
dft.pivot_table(values=['age','fare'],index=['survived','embarked','sex'],aggfunc=['min','max','mean'])
```

# date_range()

It is a function which is used to generate the sequence of date in b/w the specified limit.
  
syntax:-

>***pd.date_range(start = 'yyyy-mm-dd' ,end = 'yyyy-mm-dd')***

```python
pd.Series(pd.date_range(start = '10-01-2010', end = '05-12-2025'))
```

```python
date = pd.date_range(start = '10-01-2010', end = '05-12-2025' , freq = 'ME')
pd.Series(date)
```

```python
date = pd.date_range(start = '2002-01-11', periods = 1309 , freq = 'D')
dft['date'] = pd.Series(date)
dft
```
