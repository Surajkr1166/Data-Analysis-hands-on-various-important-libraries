```python
import numpy as np
import pandas as pd
```

# 0. Reading the  data


```python
dfa =pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv ', encoding = 'ISO8859-'   )
dfa
```

#1. Renaming the columns name.

> ***syntax : df_var.rename(columns = dict_var/dict)***

> ***use inplace = true/false  atrribute to make it permanent or***

> ***use df_var = df_var.rename(columns = dict_var/dict)***

```python
# dfa.rename(columns=({'ano':'year' , 'mes':'momth' , 'estado':'state' , 'numero':'no_of_fires' , 'encontro':'meeting_date'}),inplace = True)
dfa = dfa.rename(columns=({'ano':'year' , 'mes':'month' , 'estado':'state' , 'numero':'no_of_fires' , 'encontro':'meeting_date'}))
dfa
```

# 2. Removing the duplicates rows

> ***df_var[df_var.duplicated] - To see duplicate rows***

> ***df_var.drop_duplicates(inpalce = true/false) -To remove duplicate rows***

```python
# len(dfa[dfa.duplicated])
dfa.drop_duplicates(inplace = True)
# dfa.info()
```

# 3. Changing the values in the particular column

> ***df_var[colName].unique() - To see duplicate values uniquely***

> ***df_var[colName] = df_var[colName].map(dict_var)***

```python
# dfa['month'].unique()
d1 = {'Janeiro':'Jan', 'Fevereiro':'Feb', 'Março':'Mar', 'Abril':'Apr', 'Maio':'May', 'Junho':'Jun', 'Julho':'Jul',
       'Agosto':'Aug', 'Setembro':'Sep', 'Outubro':'Oct', 'Novembro':'Nov', 'Dezembro':'Dec'}
dfa['month'] = dfa['month'].map(d1)
dfa
```

#4. Removing unwanted data in the particular column

>***df_var[colName] = df_var[colName].str.strip('unwanted data')***

```python
dfa['no_of_fires'] = dfa['no_of_fires'].str.strip('_Fires')
dfa['no_of_fires']
```

#5. Applying in-built function on a particular column

> ***df_var[colName] = df_var[colName].str.upper()/lower()/tiltle()***

```python
dfa['month'] = dfa['month'].str.upper()
dfa
```

```python
dfa['state'] = dfa['state'].str.lower()
dfa
```

```python
dfa['state'] = dfa['state'].str.title()
dfa
```

# 6. Handling the  missing elements

1. deleting all the NAN values

    syntax :

    > ***df_var.dropna(inplace = true/false)***

    > ***df_var.dropna(subset = ['colName1','colName2',...,'colNameN'] , inplace = true/false)***

2. filling the NAN values with 0 or Avg (Numerical Col / Object Col)

    syntax :

    avg = df_var['colName'].mean()

     > ***df_var['colName'] = df_var['colName'].fillna(0/avg)***

    dfd.describe(include = object)
    
     > ***df_var['colName'] = df_var['colName'].fillna(object)***

    df_var['colName'].value_counts()

     > ***df_var['colName'] = df_var['colName'].fillna(object)***

    mrv = df_var['colName'].mode()[0]

     > ***df_var['colName'] = df_var['colName'].fillna(mrv)***

3. forward filling and backward filling

    syntax :

    > ***df_var['colName'] = df_var['colName'].ffill()***

    > ***df_var['colName'] = df_var['colName'].bfill()***


```python
# creating the dummy dataframe having name column age column bday column
# including seven rows fill the columns
dummy={'name':['a','b','c','b','a','f',None,'g'],'age':[10,11,12,13,15,None,19,21],'bday':[2,4,3,10,23,13,25,None]}
dfd=pd.DataFrame(dummy)
dfd
```

```python
# delete all null values
# dfd.dropna(inplace=True)
# dfd
```

```python
# delete particular column's null values
dfd.dropna(subset = ['age'] , inplace=True)
dfd
```

```python
# dfd['bday']= dfd['bday'].fillna(0)
# dfd

dfd['bday']= dfd['bday'].fillna(dfd['bday'].mean())
dfd
```

```python
# dfd.describe(include = object)
# dfd['name'] = dfd['name'].fillna('a')
# dfd

# dfd['name'].value_counts()
# dfd['name'] = dfd['name'].fillna('a')
# dfd

mrv = dfd['name'].mode()[0]
dfd['name'] = dfd['name'].fillna(mrv)
dfd
```

```python
# # forward filling
# dfd['age'] = dfd['age'].ffill()
# dfd

# dfd['bday'] = dfd['bday'].ffill()
# dfd


# backward filling
dfd['name'] = dfd['name'].bfill()
dfd
```

# 7. Changing the  the datatype of the columns

syntax :

> ***df_var['colName'] = df_var['colName'].astype(datatype)***



```python
# # col without null can only be converted for datatype
# dfd['age'] = dfd['age'].astype(int)
# dfd

# col without null can only be converted for datatype (multiple col using dictionary)
dfd = dfd.astype({'age' : 'int' ,'bday': 'int'})
dfd
```

# 8. Changing the order of the column

syntax :

> ***df_var = df_var[df_var.colName[list_var/list_collection]]***

```python
ind =[1,2,0]
dfd = dfd[dfd.columns[ind]]
dfd
```

# 9. Changing the particular value in a particular column

syntax :

> ***df_var['colName'] = df_var[colName].replace('oldVal','newVal')***

> ***df_var['colName'] = df_var[colName].str.replace('oldVal','newVal')***

```python
dfa['year'] = dfa['year'].replace(1998,2025)
dfa
```

```python
dfa['month'] = dfa['month'].str.replace('jan','january')
dfa
```
