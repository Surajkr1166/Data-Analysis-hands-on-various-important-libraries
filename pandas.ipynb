## 7 loc VS iloc

1. Ending Value Inclusion:
    - "**loc**": The ending value is included.
    - "**iloc**": The ending value is excluded.
2. Column Slicing:
    - "**loc**": Column slicing is possible.
    - "**iloc**": Column slicing is possible.
3. Acceptance of Column Names/Slicing:
    - "**loc**": It accepts 'col. Names' but doesn't accept column slicing in the same way "iloc" does; however, it is possible to slice using column names.
    - "**iloc**": It doesn't accept 'col. Names' but accepts column slicing using integer positions

### 7.1 loc

####syntax:-

> ***1. df_var.loc[rowSlicing,['colName1','colName2',....,'colNameN']]***

> ***2. df_var.loc[[rowindex1,rowindex2,....,rowindexN],['colName1','colName2',....,'colNameN']]***


```python
# rowSlicing using loc
dft.loc[0:500:5,['name','pclass','sex','age']]
```

```python
# row-index using loc
dft.loc[[1,56,5,27,10],['name','pclass','sex','age']]
```

### 7.2 iloc

####syntax:-

> ***1. df_var.iloc[rowSlicing,colSlicing]***

> ***2. df_var.iloc[[rowindex1,rowindex2,....,rowindexN],[colIndex1,colIndex2,....,colIndexN]]***


```python
dft.iloc[1:10:,0:11:2]
```

```python
dft.iloc[[1,10,20,30],[1,2,3,4]]
```

## 8 Boolen Filetring

This process is used to fetch the data using condition.

####Syntax:

> ***1. codition***

> ***2. df_var[condition]***

> ***3. df_var[colName][condition]***

> ***4. df_var[colName1,colName2,...,colNameN][condition]***

> ***5. df_var[(condition1) oprator{&,|} (condition2)...opr{&,|} (conditionN)]***

> ***6. df_var[colName1,colName2,...,colNameN][(condition1) opr{&,|} (condition2)...opr{&,|} (conditionN)]***

> ***7. df_var[[colName].isnull()]***

> ***8. df_var[[colName].notnull()]***

```python
dft['sex'] == 'male'
```

```python
dft[dft['sex'] == 'male']
```

```python
dft['name'][dft['sex'] == 'male']
```

```python
dft[['name','age','sex']][dft['age'] == 14.0]
```
